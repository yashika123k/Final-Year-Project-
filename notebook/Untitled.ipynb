```python
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns = ["energy","round","death","ch"]
df = pd.read_csv("../rust-sim/src/LEACH.csv",names=columns)
```

```python
df["death"] = 100 - df["death"]
```

```python

enrgy_per_round = df.groupby("round")["energy"].mean()
death_per_round = df.groupby("round")["death"].mean()
```

```python
sns.lineplot(x=enrgy_per_round.index,y=enrgy_per_round)
```

```python
sns.lineplot(x=death_per_round.index,y=death_per_round)
```
